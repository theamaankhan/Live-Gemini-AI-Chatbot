```python
import os
import time
import google.generativeai as genai
from datetime import datetime
```

```python
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY") or input("Enter your Google API key: ")
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model = genai.GenerativeModel("models/gemini-1.5-flash")
```

```python
def get_timestamp():
    now = datetime.now()
    return now.strftime("%B %d, %Y %I:%M %p")
```

```python
system_prompt = (
    f"You are Gemini, a helpful AI. The current date and time is {get_timestamp()}. "
    "Answer conversationally and use this as 'today' throughout."
)
chat = model.start_chat(history=[])

print("ðŸ¤– Gemini Chatbot (live) â€” type 'exit' to quit\n")

while True:
    user_input = input("Hi how can i help  you : ")
    if user_input.strip().lower() in ("exit", "quit", "bye"):
        print("ðŸ‘‹ Goodbye!")
        break

    timestamp = get_timestamp()
    full_input = f"{system_prompt}\nUser: {user_input} (Current time: {timestamp})"

    try:
        response = chat.send_message(full_input, stream=True)
        print("Gemini:", end=" ", flush=True)
        for chunk in response:
            print(chunk.text, end="", flush=True)
        print("\n")
    except genai.Error as e:
        print(f"Error: {e}. Reinitializing sessionâ€¦")
        time.sleep(1)
        chat = model.start_chat(history=chat.history)

print("Chat session ended.")
```
